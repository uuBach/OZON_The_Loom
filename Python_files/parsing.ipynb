```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

options = Options()
options.add_argument("--disable-blink-features=AutomationControlled")

service = Service("/Volumes/BigBox/Downloads/chromedriver-mac-arm64/chromedriver")
driver = webdriver.Chrome(service=service, options=options)

query = "пушистые штаны"
url = f"https://www.ozon.ru/search/?text={query.replace(' ', '+')}"
driver.get(url)

WebDriverWait(driver, 20).until(
    EC.presence_of_all_elements_located((By.CSS_SELECTOR, "div.qi7_24"))
)

for i in range(10):
    driver.execute_script("window.scrollBy(0, 3000);")
    time.sleep(2)

cards = driver.find_elements(By.CSS_SELECTOR, "div.qi7_24")
print("Нашёл карточек:", len(cards))

results = []
for card in cards[:100]: 
    try:
        title_el = card.find_elements(By.CSS_SELECTOR, "span.tsBody500Medium")
        price_el = card.find_elements(By.CSS_SELECTOR, "span.tsHeadline500Medium")

        title = title_el[0].text.strip() if title_el else ""
        price = price_el[0].text.strip() if price_el else ""

        if title and price:
            results.append({"Название": title, "Цена": price})
            print(title, "-", price)
    except:
        continue

df = pd.DataFrame(results)
df.to_excel("ozon_prices.xlsx", index=False)
print("✅ Сохранено в ozon_prices.xlsx")

driver.quit()
```
